```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import RobustScaler, StandardScaler


data.drop(columns=['bike_id'], inplace=True, errors='ignore')


lat_lon_cols = [
    'start_station_latitude', 'end_station_latitude',
    'start_station_longitude', 'end_station_longitude'
]
scaler_robust = RobustScaler()
data[lat_lon_cols] = scaler_robust.fit_transform(data[lat_lon_cols])


data['duration_sec'] = np.log1p(data['duration_sec'])
scaler_duration = StandardScaler()
data[['duration_sec']] = scaler_duration.fit_transform(data[['duration_sec']])


station_id_cols = ['start_station_id', 'end_station_id']
scaler_ids = StandardScaler()
data[station_id_cols] = scaler_ids.fit_transform(data[station_id_cols])


scaler_birth = StandardScaler()
data[['member_birth_year']] = scaler_birth.fit_transform(data[['member_birth_year']])

```

```python
cols_to_show = [
    'start_station_latitude', 'end_station_latitude', 
    'start_station_longitude', 'end_station_longitude',
    'duration_sec', 'start_station_id', 'end_station_id', 
    'member_birth_year'
]

data_description = data[cols_to_show].describe().T
print(data_description)
```

```python
# trying to fill missing values in my datset using for loop 

for  col in data.select_dtypes('number').columns:
    if data[col].isnull().any() == True :
        data[col] = data[col].fillna(data[col].median())

print(data.select_dtypes(include='number').isnull().sum())

print('==========================================================')


for  col in data.select_dtypes('object').columns:
    if data[col].isnull().any() == True :
        data[col] = data[col].fillna(data[col].mode())
print(data.select_dtypes(include='object').isnull().sum())
```

```python
##  still got a missing value in the member_gender, so let's explore this column more 

# print(bf.member_gender.head(15))
data['member_gender'] = data['member_gender'].fillna(data['member_gender'].mode()[0])
data['end_station_name'] = data['end_station_name'].fillna(data['end_station_name'].mode()[0])
data['start_station_name'] = data['start_station_name'].fillna(data['start_station_name'].mode()[0])
```
